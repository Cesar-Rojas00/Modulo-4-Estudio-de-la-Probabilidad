```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos de ventas
num_clientes = [0, 1, 2, 3, 4, 5]
frecuencias = [20, 30, 25, 15, 8 ,2]
```

## Calcular probabilidad

```python
total_dias = sum(frecuencias)
```

```python
# Probabilidad de numero de clientes adoptando el plan
probabilidades = [f/total_dias for f in frecuencias]
```

```python
# Mostrtar las probabilidades calculadas
for n, p in zip(num_clientes, probabilidades):
    print(f'Número de clientes: {n}, Probabilidad: {p:.2f}')
```

```python
# Grafico de barras para visualizar la distribución de probabilidad
plt.title('Distribución de la probabilidad de ventas en planes tarifarios')
plt.xlabel('Número de Clientes que contratan el plan')
plt.ylabel('Probabilidad')

# Crear los colores
colores = plt.cm.viridis(np.linspace(0, 1, len(num_clientes)))

# Crear el gráfico de barras
plt.bar(num_clientes, probabilidad, color = colores)

# Agregar valores en la parte superior de las barras
for x, y in zip(num_clientes, probabilidades):
    plt.text(x, y, f'{y:.2f}', ha = 'center', va = 'bottom')

# Ajustar los límites en el eje y
plt.ylim(0, max(probabilidades) * 1.1)
plt.show()
```
