```python
import matplotlib.pyplot as plt
import numpy as np
```

## Valor Esperado

```python
# Definir los datos de ventas
ventas = [0, 1, 2, 3, 4, 5]
dias = [54, 117, 72, 42, 12, 3]

# Calcular la probabilidad
total_ventas = sum(dias)
probabilidad = [dia/total_ventas for dia in dias]

# Calcular el valor esperado
valor_esperado = sum([venta * prob for venta, prob in zip(ventas, probabilidad)])
```

```python
# Elaborar gráfica
plt.xlabel('Número de automóviles vendidos')
plt.ylabel('Probabilidad')
plt.title('Distribución de probabilidad de ventas de automóviles en GM Motors')

colores = plt.cm.viridis(np.linspace(0, 1, len(ventas)))

plt.bar(ventas, probabilidad, color = colores)

for x, y in zip(ventas, probabilidad):
    plt.text(x, y, f'{y:.2f}', ha = 'center', va = 'bottom')

plt.axvline(x = valor_esperado, color = 'red', linestyle = '--', label = f'Valor Esperado: {valor_esperado:.2f}')

plt.legend()
plt.show()
```

## Varianza

```python
# Agregamos la varianza
varianza = sum([(venta - valor_esperado) ** 2 * prob for venta, prob in zip(ventas, probabilidad)])
```

```python
# Agregamos la línea de la varianza
plt.xlabel('Número de automóviles vendidos')
plt.ylabel('Probabilidad')
plt.title('Distribución de probabilidad de ventas de automóviles en GM Motors')

colores = plt.cm.viridis(np.linspace(0, 1, len(ventas)))

plt.bar(ventas, probabilidad, color = colores)

for x, y in zip(ventas, probabilidad):
    plt.text(x, y, f'{y:.2f}', ha = 'center', va = 'bottom')

plt.axvline(x = valor_esperado, color = 'red', linestyle = '--', label = f'Valor Esperado: {valor_esperado:.2f}')

plt.axvline(x = valor_esperado + varianza, color = 'green', linestyle = '--', label = f'Varianza +: {varianza:.2f}')
plt.axvline(x = valor_esperado - varianza, color = 'blue', linestyle = '--', label = f'Varianza -: {varianza:.2f}')

plt.legend()
plt.show()
```

## Desviación Estándar

```python
# Calculamos la desviacion estandar
desviacion_estandar = np.sqrt(varianza)
```

```python
# Agregamos la línea de desviacion estandar
plt.xlabel('Número de automóviles vendidos')
plt.ylabel('Probabilidad')
plt.title('Distribución de probabilidad de ventas de automóviles en GM Motors')

colores = plt.cm.viridis(np.linspace(0, 1, len(ventas)))

plt.bar(ventas, probabilidad, color = colores)

for x, y in zip(ventas, probabilidad):
    plt.text(x, y, f'{y:.2f}', ha = 'center', va = 'bottom')

plt.axvline(x = valor_esperado, color = 'red', linestyle = '--', label = f'Valor Esperado: {valor_esperado:.2f}')

plt.axvline(x = valor_esperado + varianza, color = 'green', linestyle = '--', label = f'Varianza +: {varianza:.2f}')
plt.axvline(x = valor_esperado - varianza, color = 'blue', linestyle = '--', label = f'Varianza -: {varianza:.2f}')

plt.axvline(x = valor_esperado + desviacion_estandar, color = 'orange', linestyle = '--', label = f'Desviación Estandar: {desviacion_estandar:.2f}')

plt.legend()
plt.show()
```
