```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Relación Lineal Positiva

```python
# Generar datos aleatorios
np.random.seed(1)
# Añadimos 100 puntos entre 0 y 10
x = np.linspace(0, 10, 100)
# Introducimos variabilidad en los datos de y
y = 2 * x + np.random.normal(0, 1, 100)
```

```python
# Representación de manera visual por diagrama de dispersión
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y', rotation = 360)
plt.title('Relación lineal positiva')

# Dividir en 4 cuadrantes
plt.axvline(x = 5, color = 'grey', linestyle = '--')
plt.axhline(y = 10, color = 'grey', linestyle = '--')

# Enumerar los cuadrantes
plt.text(0.05, 0.9, 'II', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.05, 0.1, 'III', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.9, 0.9, 'I', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.9, 0.1, 'IV', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)

plt.show()
```

## Relación lineal aproximadamente cero

```python
# Generar datos aleatorios
np.random.seed(2)
# Añadimos 100 puntos entre 0 y 10
x = np.linspace(0, 10, 100)
# Introducimos 100 datos distribuidos normalmente centrados en 0 y una desviacion estandar de 1
y = np.random.normal(0, 1, 100)
```

```python
# Representación de manera visual por diagrama de dispersión
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y', rotation = 360)
plt.title('Relación lineal aproximadamente cero')

# Dividir en 4 cuadrantes
plt.axvline(x = 5, color = 'grey', linestyle = '--')
plt.axhline(y = 0, color = 'grey', linestyle = '--')

# Enumerar los cuadrantes
plt.text(0.05, 0.9, 'II', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.05, 0.1, 'III', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.9, 0.9, 'I', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.9, 0.1, 'IV', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)

plt.show()
```

## Relación lineal negativa

```python
np.random.seed(3)
# Añadimos 100 puntos entre 0 y 10
x = np.linspace(0, 10, 100)
# Mientras x aumente y disminuye
y = -2 * x + np.random.normal(0, 1, 100)
```

```python
# Representación de manera visual por diagrama de dispersión
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y', rotation = 360)
plt.title('Relación lineal negativa')

# Dividir en 4 cuadrantes
plt.axvline(x = 5, color = 'grey', linestyle = '--')
plt.axhline(y = -10, color = 'grey', linestyle = '--')

# Enumerar los cuadrantes
plt.text(0.05, 0.9, 'II', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.05, 0.1, 'III', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.9, 0.9, 'I', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)
plt.text(0.9, 0.1, 'IV', fontsize = 14, fontweight = 'bold', transform = plt.gca().transAxes)

plt.show()
```

## Ejercicio 1

```python
# Crear dataframe con los datos de gasto en marketing y ventas mensuales
data = {
    'Gasto en marketing': [10000, 12000, 8000, 15000, 11000, 13000],
    'Ventas mensuales': [50000, 55000, 45000, 60000, 52000, 58000]
}

df = pd.DataFrame(data)
```

```python
# Configurar etiquetas y titulo
df.plot.scatter(x = 'Gasto en marketing', y = 'Ventas mensuales')
plt.xlabel('Gastos en marketing')
plt.ylabel('Ventas mensuales')
plt.title('Diagrama de dispersión: Gastos en marketing vs Ventas mensuales')

plt.show()
```

### Covarianza y Coeficiente de correlación

```python
# Calculo de covarianza (grado en que dos variables cambian juntas, un valor positivo indica que ambas tienden a aumentar juntas)
covariance = df['Gasto en marketing'].cov(df['Ventas mensuales'])
```

```python
# Calculo de coeficiente de correlacion (indica relacion positiva, negativa o cero)
coef_corr = df['Gasto en marketing'].corr(df['Ventas mensuales'])
```

```python
# Configurar etiquetas y titulo
df.plot.scatter(x = 'Gasto en marketing', y = 'Ventas mensuales')
plt.xlabel('Gastos en marketing')
plt.ylabel('Ventas mensuales')
plt.title('Diagrama de dispersión: Gastos en marketing vs Ventas mensuales')

# Mostrar covarianza en diagrama
plt.legend(['Covarianza: ' + str(round(covariance))])

# Mostrar coeficiente de correlación en diagrama
plt.annotate('Coeficiente de correlación: ' + str(round(coef_corr, 2)), xy = (11000, 50000))

plt.show()
```
